# Number of Matches per Venue


```python
sns.countplot(data=match, x='venue')
plt.xticks(rotation='vertical')
plt.title('Number of matches played at different venues')
```

```python
x = match['team1'].value_counts()
y = match['team2'].value_counts()
(x+y).plot(kind='barh')
plt.title('Number of matches played by different teams')
plt.xlabel('Number of matches')
```

```python
x=pd.DataFrame({"Winner":match['winner']}).value_counts()
print(x)
```

```python
sns.countplot(data=match,x = 'winner')
plt.xticks(rotation='vertical')
plt.title('Number of matches won by different teams')
```

```python
temp_data=match['player_of_match'].value_counts().head()
print(temp_data)
```

```python
sns.barplot(
    x=temp_data.values,
    y=temp_data.index,
)

plt.title("Top 5 MoM")
plt.xlabel("Match Count")
plt.ylabel("Player")
plt.show()
```

highest number of runs

```python
top_batsman=delivery.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values('batsman_runs', ascending=False).head(10)
top_batsman.set_index('batsman', inplace=True)
top_batsman.plot(kind='bar')
```

Bowler Who Has Given the Highest Number of Runs

```python
most_runs = delivery.groupby('bowler')['total_runs'].agg('sum').reset_index().sort_values('total_runs', ascending=False).head(10)
```

```python
most_runs.set_index('bowler', inplace=True)
most_runs.plot(kind='bar')
```

```python
mask=delivery['bowler']=='PP Chawla'
delivery[mask].groupby('batting_team')['total_runs'].agg('sum').plot(kind='bar')

# We are taking the example of PP Chawla. This bowler has given the highest number of runs in the IPL history till 2020. We summed up the total runs given by PP Chawla to the opponent team.
```

```python
# Over-wise Batting Performance of Each Team in the IPL (2008â€“2020)
delivery6=delivery[mask]
delivery6=delivery6[['batting_team','over','batsman_runs']]
x=delivery6.pivot_table(values='batsman_runs', index='batting_team', columns='over', aggfunc='count')
sns.heatmap(x, cmap='summer')
```

```python
#Dismissal Kind
sns.countplot(data=delivery, x='dismissal_kind')
plt.xticks(rotation='vertical')
```

```python
mask=delivery['bowler']=='JJ Bumrah'
mask2=delivery['batsman']=='V Kohli'
delivery[mask].groupby('batsman')['batsman_runs'].agg('count').sort_values(ascending=False)['V Kohli']
sns.countplot(data=delivery[mask & mask2], x='batsman_runs')
plt.title('Runs scored by V Kohli off JJ Bumrah')
```
