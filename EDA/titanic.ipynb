```python
df1 = pd.read_csv('test.csv')
df1.head()
```

```python
df = pd.concat([df, df1], ignore_index=True)
```

```python
df[df["Ticket"]== "CA. 2343"]
```

```python
Individual_Fare = df["Fare"] / (df["SibSp"] + df["Parch"] + 1)
df["Individual_Fare"] = Individual_Fare
```

```python
df
```

```python
Family= df['SibSp'] + df['Parch'] + 1
df['Family_Size'] = Family
```

```python
df
```

```python
def transform_family_size(numb):

    if numb == 1:
        return 'Alone'
    elif numb <= 4:
        return 'Small'
    else:
        return 'Large'
```

```python
df['Family_Size'] = df['Family_Size'].apply(transform_family_size)
```

```python
df
```

```python
pd.crosstab(df['Survived'], df['Family_Size'],normalize='columns')*100
```

```python
df['Surname'] = df['Name'].str.split(',').str[0]
```

```python
df
```

```python
title = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ').str.get(0)
df['Title'] = title
```

```python
df
```

```python
df["Title"].value_counts()
```

```python
other = ['Don.', 'Rev.', 'Dr.', 'Major.', 'Lady.', 'Sir.', 'Col.', 'Capt.', 'the', 'Jonkheer.', 'Dona.']
df['Title'] = df['Title'].replace(other, 'Other')
```
